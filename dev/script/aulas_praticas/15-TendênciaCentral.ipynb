```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics as st
```

```python
obj_data = list(np.random.normal(1_000,500,5_000))
obj_data += list(np.random.normal(500,300,2_500))
obj_data_max = max(obj_data)
obj_data_min = min(obj_data)
obj_data += list([i + obj_data_max for i in range(1, 100, 17)])
obj_data += list([i + obj_data_min for i in range(1, 100, 17)])
obj_data_mean = round(st.mean(obj_data), 2)
obj_data_median = round(st.median(obj_data), 2)
obj_data_mode = round(st.mode(obj_data), 2)
medidas_resumo = f" .-.-.- Mean: {obj_data_mean} \n ...... Mean: {obj_data_median} \n ------ Moda: {obj_data_mode}"
```

```python
plt.figure(figsize=(20,5))
sns.histplot(
    kde=True,
    element='poly',
    data=obj_data
)
plt.title("Medidas de Tendência Central")
plt.xlabel("Valores")
plt.ylabel('Contagem')
plt.text(-1000, 350, medidas_resumo, fontsize="x-large")
plt.axvline(obj_data_mean, 0, 2, color="b", lw=4, ls=":")
plt.axvline(obj_data_median, 0, 2, color="r", lw=4, ls="-.")
plt.axvline(obj_data_mode, 0, 2, color="y", lw=2.5, ls="--")
print_img = False
if(print_img):
    plt.savefig('figuras/16-Tendencia_Central_hist.png', transparent=True, bbox_inches='tight')
plt.show()
```

```python
plt.figure(figsize=(20,5))
sns.boxplot(
    x=obj_data,
    color='g'
)
if(print_img):
    plt.savefig('figuras/16-Tendencia_Central_box.png', transparent=True, bbox_inches='tight')
plt.show()
```
