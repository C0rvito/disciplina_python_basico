```python
def soma(num1, num2):
    # Função que recebe dois valores e retorna sua soma
    if type(num1) != str and type(num2) != str:
        return num1+num2
    else:
        return "Valores não são inteiros"
```

```python
print(soma(20, "5.1"))
```

```python
for _ in range(200):
    print(soma(60, 10))
```
